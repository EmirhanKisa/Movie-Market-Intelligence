```python
from ast import literal_eval

# 1. Önce o karmaşık yazıyı Python'un anlayacağı gerçek listeye çeviriyoruz
# (Eğer hata alırsan bu satırı try-except bloğuna almak gerekebilir ama genelde çalışır)
df['genres'] = df['genres'].apply(literal_eval)

# 2. Listenin içine girip sadece 'name' (isim) kısımlerını alıyoruz
# Yeni kolonumuzun adı: 'turler'
df['turler'] = df['genres'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])

# 3. Sonucu kontrol edelim (Eski hali vs Yeni hali)
print(df[['genres', 'turler']].head())
```

```python
# df["collection_name"]
print(df["turler"])
```

```python
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce').fillna(0)
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0)
df["Karımız"] = df["revenue"] - df["budget"]
```

```python
print(df["Karımız"])
```

```python
# 1. Önce hesaplama yapabilmek için sayısal verileri garantiye alalım
# (Hatalı verileri NaN yapar, 0 olan bütçeleri atarız ki ROI sonsuz çıkmasın)
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Bütçesi 1000 dolardan az olanları atalım (Veri kirliliğini önlemek için)
df_analiz = df[df['budget'] > 1000].copy()

# 2. ROI (Yatırım Getirisi) Hesapla
# Formül: (Hasılat - Bütçe) / Bütçe
# Örnek: 1 koyup 3 aldıysan ROI = 2 (yani %200 kâr)
df_analiz['roi'] = (df_analiz['revenue'] - df_analiz['budget']) / df_analiz['budget']

# 3. İŞTE SİHİRLİ KISIM: Türleri ayırıyoruz (Explode)
# Bu işlem satır sayısını artırır. Bir film 3 türe sahipse, 3 satır olur.
df_exploded = df_analiz.explode('turler')

# 4. Artık Gruplama yapabiliriz
# Her tür için Ortalama ROI ve Ortalama Bütçe'yi hesapla
tur_performansi = df_exploded.groupby('turler').agg({
    'roi': 'mean',          # Ortalama getiri
    'budget': 'mean',       # Ortalama harcanan para
    'id': 'count'           # O türde kaç film var? (Veri güvenilirliği için)
})

# 5. Sadece en az 50 film çekilmiş türleri alalım (İstatistiksel olarak anlamlı olsun)
tur_performansi = tur_performansi[tur_performansi['id'] > 50]

# 6. ROI'ye göre sırala (En kârlıdan en aza)
sonuc = tur_performansi.sort_values(by='roi', ascending=False)

# Tabloyu görelim
print(sonuc.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tasarım ayarları (Grafikler şık görünsün)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(18, 8))

# --- GRAFİK 1: ROI SIRALAMASI (Bar Chart) ---
plt.subplot(1, 2, 1) # Sol taraftaki grafik
# Renk paletini ROI değerine göre koyulaştırarak vurgu yapalım
sns.barplot(x=sonuc['roi'], y=sonuc.index, palette="viridis")

plt.title('Film Türlerine Göre Yatırım Getirisi (ROI)', fontsize=14, fontweight='bold')
plt.xlabel('ROI (Yatırım Getiri Oranı)', fontsize=12)
plt.ylabel('Film Türleri', fontsize=12)
# Mystery çubuğunun ne kadar uzun olduğuna dikkat et!

# --- GRAFİK 2: RİSK vs. ÖDÜL MATRİSİ (Scatter Plot) ---
plt.subplot(1, 2, 2) # Sağ taraftaki grafik

# Bütçeyi Milyon Dolar cinsinden gösterelim ki eksen okunabilsin
bute_milyon = sonuc['budget'] / 1_000_000

# Scatter plot çizimi
# x=Risk (Bütçe), y=Ödül (ROI), size=Film Sayısı (Daire büyüklüğü o türdeki film sayısını göstersin)
sns.scatterplot(x=bute_milyon, y=sonuc['roi'], size=sonuc['id'], sizes=(100, 1000), hue=sonuc.index, palette="deep", legend=False, alpha=0.7)

# Her noktanın üzerine türün adını yazalım
for i in range(len(sonuc)):
    plt.text(x=bute_milyon[i]+0.5, y=sonuc['roi'][i], s=sonuc.index[i], fontdict=dict(color='black', size=10, weight='bold'))

plt.title('Risk vs. Ödül Analizi (Bütçe ve ROI İlişkisi)', fontsize=14, fontweight='bold')
plt.xlabel('Ortalama Bütçe (Milyon $)', fontsize=12)
plt.ylabel('ROI (Getiri)', fontsize=12)

# Grafikleri göster
plt.tight_layout()
plt.show()
```

```python
# Veriyi temizleyelim
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# FİLTRELER (Business Logic)
# 1. Bütçesi 100.000 Dolar'dan az olanları at (Hatalı veri veya ev videosu)
# 2. Hasılatı 0 olanları at (Verisi girilmemiş)
df_analiz = df[(df['budget'] > 100000) & (df['revenue'] > 10000)].copy()

# ROI Hesapla
df_analiz['roi'] = (df_analiz['revenue'] - df_analiz['budget']) / df_analiz['budget']

# Türleri patlat
df_exploded = df_analiz.explode('turler')

# Hem Ortalama (Mean) hem Ortanca (Median) görelim
sonuc = df_exploded.groupby('turler')['roi'].agg(['mean', 'median', 'count'])

# Güvenilir olması için en az 50 film çekilmiş türleri alalım
sonuc = sonuc[sonuc['count'] > 50]

# Ortanca'ya göre sırala (Daha gerçekçi sıralama)
print(sonuc.sort_values(by='median', ascending=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tasarım Ayarları
sns.set_theme(style="whitegrid")
plt.figure(figsize=(18, 8))

# 1. Veriyi Hazırla (Hem ROI hem Bütçe MEDIAN değerleri lazım)
# budget ve revenue medianlarını da hesaplayalım
sonuc_final = df_exploded.groupby('turler').agg({
    'roi': 'median',
    'budget': 'median',
    'id': 'count'
})
# Güvenilirlik filtresi (50 filmden az olanları at)
sonuc_final = sonuc_final[sonuc_final['id'] > 50].sort_values(by='roi', ascending=False)

# --- GRAFİK 1: GERÇEKÇİ KÂRLILIK (Median ROI Bar Chart) ---
plt.subplot(1, 2, 1)
sns.barplot(x=sonuc_final['roi'], y=sonuc_final.index, palette="viridis")
plt.title('GERÇEKÇİ Yatırım Getirisi (Median ROI)', fontsize=14, fontweight='bold')
plt.xlabel('ROI (Ortanca Getiri)', fontsize=12)
plt.axvline(x=1, color='red', linestyle='--', label='Kritik Sınır (Kâr/Zarar Eşiği)') # Kırmızı çizgi ekledim
plt.legend()

# --- GRAFİK 2: RİSK ANALİZİ (Scatter Plot) ---
plt.subplot(1, 2, 2)
# Bütçeyi Milyon Dolar yapalım
bute_milyon = sonuc_final['budget'] / 1_000_000

sns.scatterplot(
    x=bute_milyon,
    y=sonuc_final['roi'],
    size=sonuc_final['id'],
    sizes=(100, 1000),
    hue=sonuc_final.index,
    palette="deep",
    legend=False,
    alpha=0.7
)

# İsimleri yazdıralım
for i in range(len(sonuc_final)):
    plt.text(
        x=bute_milyon[i]+0.5,
        y=sonuc_final['roi'][i],
        s=sonuc_final.index[i],
        fontdict=dict(color='black', size=9, weight='bold')
    )

plt.title('Risk Haritası (Median Bütçe vs ROI)', fontsize=14, fontweight='bold')
plt.xlabel('Standart Film Bütçesi (Milyon $)', fontsize=12)
plt.ylabel('ROI (Getiri)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import ast

# 1. SENİN TANIMLADIĞIN KISIM (Credits verisi)
dff = pd.read_csv('/content/credits.csv', on_bad_lines='skip', engine='python')

# 2. Movies Verisini Yükle (Para verisi burada)
# (Eğer dosya yolu farklıysa '/content/...' kısmını düzelt)
movies = pd.read_csv('/content/movies_metadata.csv', low_memory=False)

# 3. ID Temizliği (İki tabloyu yapıştırmak için şart)
# Hatalı ID'leri temizliyoruz ki merge işlemi patlamasın
dff = dff[pd.to_numeric(dff['id'], errors='coerce').notnull()]
dff['id'] = dff['id'].astype(int)

movies = movies[pd.to_numeric(movies['id'], errors='coerce').notnull()]
movies['id'] = movies['id'].astype(int)

# 4. Birleştirme (Merge)
df = movies.merge(dff, on='id')

# 5. Oyuncu İsimlerini Ayıklama (Yeni Kolon Oluşturma)
def isimleri_al(text):
    try:
        # Listeden sadece ilk 2 ismi (Başrolleri) alıyoruz
        return [kisi['name'] for kisi in ast.literal_eval(text)[:2]]
    except:
        return []

print("Oyuncular ayıklanıyor...")
df['oyuncu_isimleri'] = df['cast'].apply(isimleri_al)

# 6. Analiz: Hasılat Hesaplama
# Hasılatı sayıya çevir ve 0 olanları at
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df_analiz = df[df['revenue'] > 10000].copy()

# Listeyi patlat (Her oyuncu için satır oluştur)
df_exploded = df_analiz.explode('oyuncu_isimleri')

# Grupla ve Ortalamayı Al
sonuc = df_exploded.groupby('oyuncu_isimleri').agg({
    'revenue': 'mean',       # Ortalama Hasılat
    'id': 'count'            # Film Sayısı
})

# 7. Sonuçları Filtrele (En az 15 filmde oynamış olanlar)
starlar = sonuc[sonuc['id'] >= 15].sort_values(by='revenue', ascending=False)

# Milyon dolar cinsinden göster
starlar['revenue'] = starlar['revenue'] / 1000000

print("\n--- GİŞE KRALLARI (TOP 20) ---")
print(starlar.head(20))
```

```python
# 4. HEDEF OYUNCULARIN ANALİZİ
# Para verisini sayıya çevir ve çöp verileri at (>10.000$)
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df_analiz = df[df['revenue'] > 10000].copy()

# Sektör Ortalaması (Kıyaslama için)
genel_ortalama = df_analiz['revenue'].mean()

# Bakmak istediğin liste
hedef_oyuncular = ["Will Smith", "Johnny Depp", "Ben Stiller", "Brad Pitt", "Jim Carrey"]

print("\n" + "="*60)
print(f"SEKTÖR ORTALAMASI: ${genel_ortalama:,.0f}")
print("="*60)
print(f"{'OYUNCU':<20} | {'FİLM':<5} | {'ORT. HASILAT ($)':<20} | {'ETKİ (KAT)':<10}")
print("-" * 65)

for oyuncu in hedef_oyuncular:
    # Oyuncunun olduğu filmleri bul
    filmler = df_analiz[df_analiz['oyuncu_isimleri'].apply(lambda x: oyuncu in x)]

    if len(filmler) > 0:
        ortalama = filmler['revenue'].mean()
        kat = ortalama / genel_ortalama
        sayi = len(filmler)
        print(f"{oyuncu:<20} | {sayi:<5} | ${ortalama:,.0f}".ljust(48) + f" | {kat:.1f} KAT")
    else:
        print(f"{oyuncu:<20} | Veri Bulunamadı")

print("-" * 65)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. VERİLERİ HAZIRLAYALIM
target_actors = ["Will Smith", "Johnny Depp", "Ben Stiller", "Brad Pitt", "Jim Carrey"]
veri_listesi = []

# Sektör Ortalaması (Kıyaslama Noktası)
# Hasılatı 10k üzeri olanları baz alıyoruz
df_clean = df[df['revenue'] > 10000].copy()
genel_ortalama = df_clean['revenue'].mean() / 1000000 # Milyon Dolar cinsinden

for actor in target_actors:
    # Oyuncunun olduğu filmleri filtrele
    # (Daha önce oluşturduğumuz 'oyuncu_isimleri' kolonunu kullanıyoruz)
    filmler = df_clean[df_clean['oyuncu_isimleri'].apply(lambda x: actor in x)]

    if len(filmler) > 0:
        ort_hasilat = filmler['revenue'].mean() / 1000000 # Milyon Dolar yap
        film_sayisi = len(filmler)
        veri_listesi.append({
            'Oyuncu': actor,
            'Ortalama Hasılat (Milyon $)': ort_hasilat,
            'Film Sayısı': film_sayisi
        })

# Çizim için tablo yap
plot_df = pd.DataFrame(veri_listesi).sort_values(by='Ortalama Hasılat (Milyon $)', ascending=False)

# 2. GRAFİĞİ ÇİZELİM
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid") # Arka planı ızgaralı yap, okuması kolay olsun

# Sütun Grafiği
ax = sns.barplot(
    x='Oyuncu',
    y='Ortalama Hasılat (Milyon $)',
    data=plot_df,
    palette='viridis' # Renk paleti
)

# 3. REFERANS ÇİZGİSİ (Sektör Ortalaması)
plt.axhline(y=genel_ortalama, color='red', linestyle='--', linewidth=2, label=f'Sektör Ortalaması ({genel_ortalama:.1f}M $)')

# 4. DETAYLAR VE ETİKETLER
plt.title('GİŞE SAVAŞLARI: Hangi Oyuncu Ne Kadar Kazandırıyor?', fontsize=16, fontweight='bold')
plt.ylabel('Ortalama Film Hasılatı (Milyon $)', fontsize=12)
plt.xlabel('Oyuncular', fontsize=12)
plt.legend()

# Sütunların üzerine tam rakamları yazalım
for p in ax.patches:
    ax.annotate(f'${p.get_height():.1f}M',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points',
                fontweight='bold')

plt.show()
```
